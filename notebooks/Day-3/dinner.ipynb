```python
import pandas as pd
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
import os

# Load test & sample submission
df = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/train.csv")  # Assuming full training data is here
test_df = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/test.csv")
submission_template = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/sample_solution.csv")
test_ids = test_df['ID']
test_df_features = test_df.drop(columns=['ID'])

# Define model hyperparameter grids
gpr_hyperparams = [
    {'kernel': C(1.0) * RBF(length_scale=2.0), 'n_restarts_optimizer': 10, 'alpha': 1e-6},
    {'kernel': C(0.5) * RBF(length_scale=2.0), 'n_restarts_optimizer': 7, 'alpha': 1e-5}
]

rf_hyperparams = [
    {'n_estimators': 200, 'max_depth': 12, 'min_samples_leaf': 1, 'max_features': None},
    {'n_estimators': 300, 'max_depth': 16, 'min_samples_leaf': 2, 'max_features': 'sqrt'}
]

elasticnet_hyperparams = [
    {'alpha': 2.0, 'l1_ratio': 0.1},
    {'alpha': 1.0, 'l1_ratio': 0.5}
]

svr_poly_hyperparams = [
    {'C': 1.0, 'epsilon': 0.1, 'degree': 3, 'coef0': 1},
    {'C': 10.0, 'epsilon': 0.01, 'degree': 2, 'coef0': 0.5}
]

# Mapping properties to model types
property_model_map = {
    'BlendProperty1': ('GPR', gpr_hyperparams),
    'BlendProperty2': ('GPR', gpr_hyperparams),
    'BlendProperty3': ('ElasticNet', elasticnet_hyperparams),
    'BlendProperty4': ('GPR', gpr_hyperparams),
    'BlendProperty5': ('RF', rf_hyperparams),
    'BlendProperty6': ('GPR', gpr_hyperparams),
    'BlendProperty7': ('SVR_Poly', svr_poly_hyperparams),
    'BlendProperty8': ('ElasticNet', elasticnet_hyperparams),
    'BlendProperty9': ('ElasticNet', elasticnet_hyperparams),
    'BlendProperty10': ('GPR', gpr_hyperparams)  # Replace NN for now
}

# Run loop and generate one CSV per combination
for i in range(1, 11):
    property_name = f'BlendProperty{i}'
    model_type, hyperparam_list = property_model_map[property_name]

    print(f"\nðŸ” Testing hyperparameters for {property_name} using {model_type}...\n")

    # Define features
    features = ['Component1_fraction', 'Component2_fraction', 'Component3_fraction',
                'Component4_fraction', 'Component5_fraction'] + \
               [f'Component{j}_Property{i}' for j in range(1, 6)]

    X_train = df[features]
    y_train = df[property_name]
    X_test = test_df_features[features]

    for idx, hyperparams in enumerate(hyperparam_list):
        print(f"â†’ Config {idx} for {model_type}")

        if model_type == 'GPR':
            model = make_pipeline(
                StandardScaler(),
                GaussianProcessRegressor(**hyperparams, random_state=42)
            )
        elif model_type == 'RF':
            model = RandomForestRegressor(**hyperparams, random_state=42, n_jobs=-1)
        elif model_type == 'ElasticNet':
            model = ElasticNet(**hyperparams, random_state=42, max_iter=10000)
        elif model_type == 'SVR_Poly':
            model = make_pipeline(
                StandardScaler(),
                SVR(kernel='poly', **hyperparams)
            )

        # Fit and predict
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)

        # Save to submission file
        submission_df = submission_template.copy()
        submission_df[property_name] = predictions
        filename = f"submission_{property_name}_{model_type}_config{idx}.csv"
        submission_df.to_csv(filename, index=False)
        print(f"âœ… Saved: {filename}")
```

```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import os

# Load data
df = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/train.csv")
test_df = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/test.csv")
submission_template = pd.read_csv("/Users/MacbookPro/LocalStorage/Developer/ShellAi/dataset/sample_solution.csv")
test_ids = test_df['ID']
test_df_features = test_df.drop(columns=['ID'])

# Define default models (unchanged)
def get_default_model(property_name, input_dim):
    final_model_info = {
        'BlendProperty1': ('Gaussian_Process', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(length_scale=2.0), n_restarts_optimizer=5, random_state=42))),
        'BlendProperty2': ('Gaussian_Process', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(length_scale=2.0), n_restarts_optimizer=5, random_state=42))),
        'BlendProperty3': ('ElasticNet', ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)),
        'BlendProperty4': ('Gaussian_Process', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(length_scale=2.0), n_restarts_optimizer=5, random_state=42))),
        'BlendProperty5': ('Random_Forest', RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)),
        'BlendProperty6': ('Gaussian_Process', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(length_scale=2.0), n_restarts_optimizer=5, random_state=42))),
        'BlendProperty7': ('SVR_Poly', make_pipeline(StandardScaler(), SVR(kernel='poly', C=1.0, epsilon=0.1))),
        'BlendProperty8': ('ElasticNet', ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)),
        'BlendProperty9': ('ElasticNet', ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)),
        'BlendProperty10': ('Neural_Network', Sequential([Dense(64, activation='relu', input_shape=(input_dim,)), Dropout(0.2), Dense(64, activation='relu'), Dense(1)]))
    }
    return final_model_info[property_name]

# Only vary one hyperparameter at a time
single_param_variants = {
    'BlendProperty1': [  # GPR
        ('length_scale_1.5', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(1.5), n_restarts_optimizer=5, random_state=42))),
        ('length_scale_3.0', make_pipeline(StandardScaler(), GaussianProcessRegressor(kernel=C(1.0, (1e-3, 1e3)) * RBF(3.0), n_restarts_optimizer=5, random_state=42))),
    ],
    'BlendProperty5': [  # RF
        ('n_estimators_200', RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1)),
        ('max_depth_15', RandomForestRegressor(n_estimators=100, max_depth=15, random_state=42, n_jobs=-1)),
    ],
    'BlendProperty3': [  # ElasticNet
        ('alpha_0.5', ElasticNet(alpha=0.5, l1_ratio=0.5, random_state=42)),
        ('l1_ratio_0.2', ElasticNet(alpha=1.0, l1_ratio=0.2, random_state=42)),
    ]
}

# Submission generation
for i in range(1, 11):
    property_name = f'BlendProperty{i}'
    features = ['Component1_fraction', 'Component2_fraction', 'Component3_fraction',
                'Component4_fraction', 'Component5_fraction'] + \
               [f'Component{j}_Property{i}' for j in range(1, 6)]

    X_train = df[features]
    y_train = df[property_name]
    X_test = test_df_features[features]

    # Default model
    model_name, default_model = get_default_model(property_name, input_dim=X_train.shape[1])
    default_model.fit(X_train, y_train)
    default_preds = default_model.predict(X_test)
    default_submission = submission_template.copy()
    default_submission[property_name] = default_preds
    default_submission.to_csv(f"submission_{property_name}_{model_name}_default.csv", index=False)
    print(f"âœ… Saved default for {property_name}")

    # Custom hyperparameter variants (if any)
    if property_name in single_param_variants:
        for variant_name, variant_model in single_param_variants[property_name]:
            variant_model.fit(X_train, y_train)
            preds = variant_model.predict(X_test)
            submission = submission_template.copy()
            submission[property_name] = preds
            filename = f"submission_{property_name}_{model_name}_{variant_name}.csv"
            submission.to_csv(filename, index=False)
            print(f"âœ… Saved {filename}")
```
